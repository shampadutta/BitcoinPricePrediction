# Bitcoin Price Prediction

```
# Importing some required library to complete the Week-1 task
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
```

```
# Read the dataset in CSV format
df = pd.read_csv('https://raw.githubusercontent.com/WidhyaOrg/datasets/master/bitcoin_dataset.csv')
df.head() # check top few rows of the dataset (by default top 5 rows)

#overview of the dataset
```

```
# Microtask - 2

df['btc_market_price'][1023]  # get the value at index 1024 in btc_market_price column

# index starts from 0
# we can use alternative method as well: df.btc_market_price[1023]
```

```
# Microtask - 3
df.corr()
```

```
# Microtask - 4

# Answer of Microtask - 4 is fill na

df.isna().any() # check which column has NaN values.

# True shows, those columns have NaN values

# To handle these NaN values, it depends on case-study (use cases)
# Here, we are going to replace the NaN values with mean of the column values

df = df.fillna(df.mean())
```

```
# Microtask - 5

x = df.drop(['btc_market_price','Date'], axis=1) # taking all the feature column except date and btc_market_price
y = df[['btc_market_price']] # output response

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3) # spilting dataset into training and testing with 70% and 30%

linear_reg = LinearRegression() # linear regression instance

linear_reg.fit(x_train,y_train) # training the model

y_predict = linear_reg.predict(x_test) # testing the model on test data
```

```
# calculate mean squared error

mean_squared_error(y_test,y_predict)
```
